# ICICI BANK DATA

```python
# read NYSE ICICIBANK data
for i in range(10, 22):
    path_temp = 'DM stock dataset/ICICIBANK NYSE/IBN-20' + str(i) + '.csv'
    exec("df_icici_nyse_" + str(i) + " = pd.read_csv(path_temp)")
    exec("df_icici_nyse_" + str(i) + "['Close_Open'] = df_icici_nyse_" + str(i) + "['Close'] - df_icici_nyse_" + str(i) + "['Open']")
    exec("df_icici_nyse_" + str(i) + "['Close_Open_ratio'] = df_icici_nyse_" + str(i) + "['Close'] / df_icici_nyse_" + str(i) + "['Open']")
    exec("df_icici_nyse_" + str(i) + "['High_Low_ratio'] = df_icici_nyse_" + str(i) + "['High'] / df_icici_nyse_" + str(i) + "['Low']")
    exec("df_icici_nyse_" + str(i) + " = df_icici_nyse_" + str(i) + ".add_prefix('nyse_')")

# read NSE ICICIBANK data    
for i in range(10, 22):
    path_temp = 'DM stock dataset/ICICIBANK NSE/ICICIBANK.NS-20' + str(i) + '.csv'
    exec("df_icici_nse_" + str(i) + " = pd.read_csv(path_temp)")
    exec("df_icici_nse_" + str(i) + "['Close_Open'] = df_icici_nse_" + str(i) + "['Close'] - df_icici_nse_" + str(i) + "['Open']")
    exec("df_icici_nse_" + str(i) + "['Close_Open_ratio'] = df_icici_nse_" + str(i) + "['Close'] / df_icici_nse_" + str(i) + "['Open']")
    exec("df_icici_nse_" + str(i) + "['High_Low_ratio'] = df_icici_nse_" + str(i) + "['High'] / df_icici_nse_" + str(i) + "['Low']")
    exec("df_icici_nse_" + str(i) + "['Open_preClose'] = 0")
    exec("df_icici_nse_" + str(i) + "['Close_Open_sign'] = 0")
    exec("df_icici_nse_" + str(i) + "['Open_preClose_sign'] = 0")
    exec("df_icici_nse_" + str(i) + " = df_icici_nse_" + str(i) + ".add_prefix('nse_')")

# bind NSE and NYSE data
for i in range(10, 22):
    exec("df_icici_" + str(i) + " = pd.merge(df_icici_nse_" + str(i) + 
         ", df_icici_nyse_" + str(i) + ", left_on='nse_Date', right_on='nyse_Date')")

# bind all data
df_icici = pd.concat([df_icici_10,df_icici_11,df_icici_12,df_icici_13,df_icici_14,df_icici_15,\
                     df_icici_16,df_icici_17,df_icici_18,df_icici_19,df_icici_20,df_icici_21],ignore_index=True)

df_icici = pd.merge(df_icici, df_nasd, left_on='nse_Date', right_on='nasd_Date')
```

```python
# add new column having Open - previous day's Close in NSE
for i in df_icici.index:
    df_icici['nse_Open_preClose'].iloc[i] = df_icici['nse_Open'].iloc[i] - df_icici['nse_Close'].iloc[i-1]
    if df_icici['nse_Close_Open'].iloc[i] > 0:
        df_icici['nse_Close_Open_sign'].iloc[i] = 1
    if df_icici['nse_Open_preClose'].iloc[i] > 0:
        df_icici['nse_Open_preClose_sign'].iloc[i] = 1

df_icici['nse_Open_preClose'].iloc[0] = 0
df_icici['nse_Open_preClose_sign'].iloc[0] = 0
```

```python
# count the number of nse_Open_preClose and nyse_Close_Open moving the same direction

n_same = 0
n_opposite = 0
n_zero = 0
for i in df_icici.index - 1:
    if df_icici['nse_Open_preClose'].iloc[i+1] * df_icici['nyse_Close_Open'].iloc[i] > 0:
        n_same += 1
    elif df_icici['nse_Open_preClose'].iloc[i+1] * df_icici['nyse_Close_Open'].iloc[i] < 0:
        n_opposite += 1
    else:
        n_zero += 1
        
print("same direction :", n_same, ", opposite direction :", n_opposite, ", zero :", n_zero)
print("rate: ", n_same/(n_same+n_opposite+n_zero))
```

```python
# count the number of nse_Close_Open and the previous day's nyse_Close_Open moving the same direction

n_same = 0
n_opposite = 0
n_zero = 0
for i in df_icici.index - 1:
    if df_icici['nse_Close_Open'].iloc[i+1] * df_icici['nyse_Close_Open'].iloc[i] > 0:
        n_same += 1
    elif df_icici['nse_Close_Open'].iloc[i+1] * df_icici['nyse_Close_Open'].iloc[i] < 0:
        n_opposite += 1
    else:
        n_zero += 1
        
print("same direction :", n_same, ", opposite direction :", n_opposite, ", zero :", n_zero)
print("rate: ", n_same/(n_same+n_opposite+n_zero))
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, classification_report
```

```python
df_infy.columns
```

```python
DFy = df_infy['nse_Close_Open_sign'].iloc[1:].reset_index()
DFx1 = df_infy['nse_Open_preClose_ratio'].iloc[1:].reset_index()
DFx2 = df_infy[['nyse_Close_Open_ratio','nyse_High_Low_ratio','nasd_Close_Open_ratio','nasd_High_Low_ratio']].iloc[:-1]
DF = pd.merge(DFy,DFx1,left_index=True,right_index=True)
DF = pd.merge(DF,DFx2,left_index=True,right_index=True)
DF = DF.drop(['index_x','index_y'],axis=1)
DF = DF.drop(0)
```

```python
DF
```

```python
pp = sns.pairplot(DF, diag_kind='kde')
pp.fig.set_size_inches(10,8)
#pp.fig.suptitle('Fig.7: Pair plot of the numerical variables colored by Y', x=0.5, y=1)
plt.show()
```

# Split the data into train and test (7:3)

```python
x = DF.iloc[:,1:]
y = DF.iloc[:,0]

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=0)
```

```python
model=LogisticRegression(penalty='none', max_iter=1000)
model_fit = model.fit(x_train, y_train)
model_fit_score = model.score(x_train, y_train)
```

```python
# fitted y and cofusion matrix
y_fit = model.predict(x_train)
cm_train = confusion_matrix(y_train, y_fit)
print(cm_train)

# misclasification rate of training sets
mr_train = (cm_train[0][1]+cm_train[1][0])/sum(sum(cm_train))
print('The misclassification rate of training set is ', str(mr_train), '.')
```

```python
# predicted y and confusion matrix
y_predict = model.predict(x_test)
cm_test = confusion_matrix(y_test, y_predict)
print(cm_test)

# misclasification rate of testing sets
mr_test = (cm_test[0][1]+cm_test[1][0])/sum(sum(cm_test))
print('The misclassification rate of testing set is ', str(mr_test), '.')
print(classification_report(y_test, y_predict))
```
